```python
# impor library standard
import numpy as np
import pandas as pd

from statsmodels.stats.proportion import power_proportions_2indep, test_proportions_2indep
from statsmodels.stats.weightstats import ttest_ind
from statsmodels.stats.power import tt_ind_solve_power
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ab = pd.read_csv('abcd.csv', date_parser='date')
ab
```

```python
index = 0
ab.loc[index, 'total']
```

```python
power = []
for index in range(23):
    conversion_rate = ab.loc[index, 'treat_convrate']
    total_control = ab.loc[index, 'control']
    perbandingan = ab.loc[index, 'ratio']
    power_count = power_proportions_2indep(diff=ab.loc[index, 'treat_convrate'] - ab.loc[index, 'control_convrate'],
                                           prop2=conversion_rate, 
                                           nobs1=total_control,
                                           ratio=perbandingan,
                                           alpha=0.05,
                                           alternative='two-sided',
                                           return_results=False)
    power.append(power_count)
    
print(power)
```
