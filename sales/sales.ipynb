```python
state = {'AC':'Acre', 'AL':'Alagoas', 'AP':'Amapá', 'AM':'Amazonas', 'BA':'Bahia', 'CE':'Ceará', 
 'DF':'Distrito Federal', 'ES':'Espírito Santo', 'GO':'Goiás', 'MA':'Maranhão', 'MT':'MatoGrosso', 
 'MS':'MatoGrosso do Sul', 'MG':'Minas Gerais', 'PA':'Pará', 'PB':'Paraíba', 'PR':'Paraná',
 'PE':'Pernambuco', 'PI':'Piauí', 'RJ':'Rio de Janeiro', 'RN':'Rio Grande do Norte', 'RS':'Rio Grande do Sul',
 'RO':'Rondônia', 'RR':'Roraima', 'SC':'Santa Catarina', 'SP':'São Paulo', 'SE':'Sergipe', 'TO':'Tocantins'
}

region = {'AC':'North', 'AL':'Northeast', 'AP':'North', 'AM':'North', 'BA':'Northeast', 'CE':'Northeast', 
 'DF':'Center West', 'ES':'Southeast', 'GO':'Center West', 'MA':'Northeast', 'MT':'Center West', 
 'MS':'Center West', 'MG':'Southeast', 'PA':'North', 'PB':'North', 'PR':'South',
 'PE':'Northeast', 'PI':'Northeast', 'RJ':'Southeast', 'RN':'Northeast', 'RS':'South',
 'RO':'North', 'RR':'North', 'SC':'South', 'SP':'Southeast', 'SE':'Northeast', 'TO':'North'
}

def state_(s):
    return state[s]

def region_(r):
    return region[r]

def country_():
    return 'Brazil'

def clean_customers(data):
    
    data = data.assign(region=(data
                                   .customer_state
                                   .apply(lambda x_ : region_(x_))
                              ),
                       country=(data
                                    .customer_state
                                    .apply(lambda x_ : country_())
                               )
                      )
    data.customer_state = (data
                               .customer_state
                               .replace(to_replace=state))
    return data
```

```python

def clean_product(data):
    num = data.columns[2:].tolist()
    
    data.product_category_name = data.product_category_name.fillna('unknown').str.replace("_", " ")
    for i in num:
        data[i] = data[i].fillna(0)
        
    return data
        
prod = prod.pipe(clean_product)
```

# order_reviews

```python

def clean_reviews(data):
    col = ['review_comment_title', 'review_comment_message']
    idx_title = [90, 10283, 29802, 30940, 32174, 43344, 70538] # * 
    idx_msg = [303, 2336, 33630, 45807, 62061, 71451, 81586, 85867] # . , #
    msg_null = [10001,11339,11798,11892,26041,26985,30049,36737,38581,39584,
                 40064,41031,45138,55542,59880,60781,60839,68879,75383,81835,
                 82938,85534,87150,88653,90275,92751,93389]
    title_null = [49962, 93916]
    dt= ['review_creation_date', 'review_answer_timestamp']
    
    for i in dt:
        data[i] = pd.to_datetime(data[i])
    
    
    for i in col:
        data[i] = data[i].fillna('-')
        index = data[data[i].apply(lambda x: x.isdigit())].index.tolist()
        for idx in index:
            data.loc[idx, i] = '-'
        
    data.loc[idx_title, 'review_comment_title'] = (data
                                                       .loc[idx_title, 'review_comment_title']
                                                       .str
                                                       .replace("*", "", regex=False)
                                                       .str
                                                       .strip()
                                                  )
    
    data.loc[idx_msg, 'review_comment_message'] = (data
                                                       .loc[idx_msg, 'review_comment_message']
                                                       .str
                                                       .replace("," , "", regex=False)
                                                       .str
                                                       .replace("." , "", regex=False)
                                                       .str
                                                       .replace("#" , "", regex=False)
                                                       .str
                                                       .strip()
                                                  )
    data.loc[data.review_comment_title.str.contains('^[.,?/*#@]'), 'review_comment_title'] = '-'
    data.loc[data.review_comment_message.str.contains('^[.,?/*#@]'), 'review_comment_message'] = '-'
    data.loc[:, 'review_comment_message'] = (data
                                                 .loc[:, 'review_comment_message']
                                                 .str
                                                 .replace("\r","")
                                                 .str
                                                 .replace("\n", "")
                                                 .str
                                                 .replace("\r\n", "")
                                                 .str
                                                 .strip()
                                            )
    for i in msg_null:
        data.loc[i, 'review_comment_message'] = '-'
    for i in title_null:
        data.loc[i, 'review_comment_title'] = '-'
        
    return data
    
```
